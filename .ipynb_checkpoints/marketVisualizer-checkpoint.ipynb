## Market Visualizer Notebook
#
# **Objective:**  
# This notebook loads the cleaned CSV files for BTC/USDT, ETH/USDT, and ADA/USDT, performs additional processing (e.g., calculating percentage changes and movement classifications), and generates professional visualizations.
#
# **Structure:**
# - **Data Loading & Preprocessing:** Load CSVs, convert timestamps, compute `pct_change` and `movement`.
# - **Time Series Visualizations:** Plot closing prices and percentage changes.
# - **Statistical Visualizations:** Create movement frequency distributions, correlation matrix, and rolling volatility plots.
# - **Technical Analysis Visualizations:** Generate candlestick charts, Bollinger Bands overlays, Renko charts, and RSI subplots using mplfinance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Enable inline plotting for Jupyter Notebook
%matplotlib inline
```

#
## Data Loading and Preprocessing
#
# Load the cleaned CSV files for each symbol, convert timestamps, and compute additional columns:
# - **pct_change:** Percentage change in the closing price from the previous minute.
# - **movement:** Categorical label based on `pct_change` ("up", "down", "flat").


```python
# Define file paths (adjust paths if necessary)
files = {
    'BTC/USDT': 'binance_klines_BTCUSDT_2years.csv',
    'ETH/USDT': 'binance_klines_ETHUSDT_2years.csv',
    'ADA/USDT': 'binance_klines_ADAUSDT_2years.csv'
}

# Dictionary to hold DataFrames for each symbol
dfs = {}

for symbol, filepath in files.items():
    df = pd.read_csv(filepath)
    # Convert the 'timestamp' column to datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S')
    # Ensure the data is sorted by time
    df = df.sort_values('timestamp').reset_index(drop=True)
    dfs[symbol] = df
    print(f"{symbol} data shape: {df.shape}")
```

```python
for symbol, df in dfs.items():
    # Calculate the percentage change in the 'close' price
    df['pct_change'] = df['close'].pct_change() * 100
    
    # Create a new column to label the movement as 'up', 'down', or 'flat'
    df['movement'] = df['pct_change'].apply(
        lambda x: 'up' if x > 0 else ('down' if x < 0 else 'flat')
    )
    
    print(f"Data for {symbol} after adding percentage change and movement columns:")
    print(df[['timestamp', 'close', 'pct_change', 'movement']].head())
```

#
## Visualization 1: Time Series Plot of Closing Prices
#
# This plot shows the closing prices for BTC/USDT, ETH/USDT, and ADA/USDT over the entire data period.

```python
plt.figure(figsize=(14, 8))
for symbol, df in dfs.items():
    plt.plot(df['timestamp'], df['close'], label=symbol)

plt.xlabel('Timestamp')
plt.ylabel('Close Price (USDT)')
plt.title('1-Minute Close Price Over 2 Years')
plt.legend()
plt.tight_layout()
plt.show()
```

#
## Visualization 2: Time Series Plot of Percentage Changes
#
# The following chart displays the percentage change in the close price for each symbol.

```python
plt.figure(figsize=(14, 8))
for symbol, df in dfs.items():
    if 'pct_change' not in df.columns:
        # Calculate pct_change if not already present
        df['pct_change'] = df['close'].pct_change() * 100
    plt.plot(df['timestamp'], df['pct_change'], label=symbol)
plt.xlabel('Timestamp')
plt.ylabel('Percentage Change (%)')
plt.title('Percentage Change in Close Price Over 2 Years')
plt.legend()
plt.tight_layout()
plt.show()
```

#
## Visualization 3: Movement Frequency Distribution
#
# These count plots show the frequency of "up," "down," and "flat" movements for each asset.
#

```python
for symbol, df in dfs.items():
    if 'movement' not in df.columns:
        # Create the movement column if not already present
        df['movement'] = df['pct_change'].apply(lambda x: 'up' if x > 0 else ('down' if x < 0 else 'flat'))
    plt.figure(figsize=(8, 4))
    sns.countplot(x='movement', data=df, order=['up', 'flat', 'down'])
    plt.title(f'Movement Frequency for {symbol}')
    plt.xlabel('Movement')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

#
## Visualization 4: Correlation Matrix of Percentage Changes
#
# Merge the percentage change columns for each symbol on the timestamp to compute the correlation matrix.

```python
merged_pct_df = None
for symbol, df in dfs.items():
    temp = df[['timestamp', 'pct_change']].copy()
    temp.rename(columns={'pct_change': symbol}, inplace=True)
    if merged_pct_df is None:
        merged_pct_df = temp
    else:
        merged_pct_df = pd.merge(merged_pct_df, temp, on='timestamp', how='inner')

print("Merged percentage change data shape:", merged_pct_df.shape)
merged_pct_df.head()
```

```python
# Set 'timestamp' as the index for correlation computation
merged_pct_df.set_index('timestamp', inplace=True)
corr_pct = merged_pct_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr_pct, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Percentage Changes')
plt.tight_layout()
plt.show()
```

#
## Visualization 5: Rolling Volatility of Percentage Changes
#
# Compute and plot a 60-minute rolling standard deviation (volatility) for each asset.

```python
plt.figure(figsize=(14, 8))
for symbol, df in dfs.items():
    # Ensure 'timestamp' is the index for resampling if needed
    df.set_index('timestamp', inplace=True)
    rolling_vol = df['pct_change'].rolling(window=60).std()  # 60-minute rolling window
    plt.plot(rolling_vol.index, rolling_vol, label=symbol)
    # Reset index to original for further visualizations
    df.reset_index(inplace=True)
plt.xlabel('Timestamp')
plt.ylabel('Rolling Std of Percentage Change')
plt.title('60-Minute Rolling Volatility of Percentage Changes')
plt.legend()
plt.tight_layout()
plt.show()
```

#
## Visualization 6: Candlestick Chart with mplfinance
#
# In this section, we use mplfinance to create a candlestick chart for BTC/USDT.
# We first convert the DataFrame to have a DatetimeIndex (if not already), select a recent time window,
# and then plot the candlestick chart along with volume and moving averages.

#
### 6.1 Candlestick Chart with Moving Averages
#
# This chart shows a candlestick plot for BTC/USDT (last 7 days) with moving averages.

```python
import mplfinance as mpf

# Choose a symbol to visualize candlestick data (e.g., BTC/USDT)
btc_df = dfs['BTC/USDT'].copy()

# Ensure 'timestamp' is the index and is a DatetimeIndex
btc_df.set_index('timestamp', inplace=True)
btc_df.sort_index(inplace=True)

btc_recent = btc_df.loc[btc_df.index >= btc_df.index[-1] - pd.Timedelta(days=7)]

# Plot the candlestick chart with volume and moving averages (3, 6, and 9 periods)
mpf.plot(btc_recent,
         type='candle',
         volume=True,
         mav=(3, 6, 9),
         title="BTC/USDT Candlestick Chart (Last 7 Days)",
         style='charles')
```

#
### 6.2 Candlestick Chart with Bollinger Bands
#
# Here we overlay Bollinger Bands on a BTC/USDT candlestick chart over the past 14 days.

```python
import numpy as np
import mplfinance as mpf

# Select BTC data and subset for the last 14 days
btc_df = dfs['BTC/USDT'].copy()
btc_df.set_index('timestamp', inplace=True)
btc_df.sort_index(inplace=True)
btc_recent = btc_df.loc[btc_df.index >= btc_df.index[-1] - pd.Timedelta(days=14)]

# Calculate Bollinger Bands
window = 20
sma = btc_recent['close'].rolling(window=window).mean()
rstd = btc_recent['close'].rolling(window=window).std()
upper_band = sma + (2 * rstd)
lower_band = sma - (2 * rstd)

# Create addplots for Bollinger Bands
apds = [
    mpf.make_addplot(sma, color='blue', width=1.0, linestyle='--', ylabel='SMA'),
    mpf.make_addplot(upper_band, color='green', width=0.8, linestyle='--'),
    mpf.make_addplot(lower_band, color='red', width=0.8, linestyle='--')
]

mpf.plot(btc_recent,
         type='candle',
         volume=True,
         addplot=apds,
         title="BTC/USDT Candlestick with Bollinger Bands (Last 14 Days)",
         style='charles',
         mav=(3,6,9))
```

#

```python
# Prepare ETH data
eth_df = dfs['ETH/USDT'].copy()
eth_df.set_index('timestamp', inplace=True)
eth_df.sort_index(inplace=True)
eth_recent = eth_df.loc[eth_df.index >= eth_df.index[-1] - pd.Timedelta(days=30)]

# Plot the Renko chart (Note: mplfinance uses OHLC data but plots Renko based on price moves)
mpf.plot(eth_recent, 
         type='renko', 
         title="ETH/USDT Renko Chart (Last 30 Days)",
         style='default')
```

#
### 6.4 ETH Candlestick Chart with RSI Subplot
#
# This chart shows a candlestick chart for ETH/USDT (last 14 days) with an RSI subplot.

```python
def compute_rsi(series, period=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period, min_periods=period).mean()
    avg_loss = loss.rolling(window=period, min_periods=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Prepare ETH data and subset for the last 14 days
eth_df = dfs['ETH/USDT'].copy()
eth_df.set_index('timestamp', inplace=True)
eth_df.sort_index(inplace=True)
eth_recent = eth_df.loc[eth_df.index >= eth_df.index[-1] - pd.Timedelta(days=14)].copy()

# Compute RSI and add as a new column
eth_recent['RSI'] = compute_rsi(eth_recent['close'], period=14)

# Create an addplot for the RSI subplot. Panel 1 indicates a separate subplot below the main chart.
rsi_addplot = mpf.make_addplot(eth_recent['RSI'], panel=1, color='purple', ylabel='RSI')

mpf.plot(eth_recent,
         type='candle',
         volume=True,
         addplot=rsi_addplot,
         title="ETH/USDT Candlestick Chart with RSI (Last 14 Days)",
         style='yahoo',
         mav=(5,10),
         panel_ratios=(3,1))  # Ratio for main chart panel and RSI panel
```

#
### 6.5 ADA Line Chart with Percentage Change Overlay
#
# This visualization plots a line chart for ADA/USDT (last 30 days) with an overlay of the percentage change on a secondary y-axis.

```python
import matplotlib.ticker as mticker

# Prepare ADA data
ada_df = dfs['ADA/USDT'].copy()
ada_df.set_index('timestamp', inplace=True)
ada_df.sort_index(inplace=True)
ada_recent = ada_df.loc[ada_df.index >= ada_df.index[-1] - pd.Timedelta(days=30)].copy()

# Ensure the pct_change column exists
if 'pct_change' not in ada_recent.columns:
    ada_recent['pct_change'] = ada_recent['close'].pct_change() * 100

fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot the ADA closing price on the primary y-axis
ax1.plot(ada_recent.index, ada_recent['close'], color='blue', label='Close Price')
ax1.set_xlabel('Timestamp')
ax1.set_ylabel('Close Price (USDT)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis for percentage change
ax2 = ax1.twinx()
ax2.plot(ada_recent.index, ada_recent['pct_change'], color='orange', label='Percentage Change (%)', alpha=0.7)
ax2.set_ylabel('Percentage Change (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
ax2.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.2f'))

plt.title("ADA/USDT Line Chart with Percentage Change Overlay (Last 30 Days)")
fig.tight_layout()
plt.legend(loc='upper left')
plt.show()
```
